```python
#listings sorted by attribute  top 10 
listings.sort_values('number_of_reviews', ascending = False).iloc[0:10]
```

```python
def popular(s):
    """Recieves s, a Pandas Series, contaning number_of_views in order of highest count to lowest count.
    Return the highest number of reviews."""
    return s.iloc[0]  
    
```

```python
#sorted by number_of_reviews in each neighbourhood_group

sorted_listings = listings.sort_values('number_of_reviews', ascending = False).groupby(['neighbourhood_group']).agg(popular)
```

```python
sorted_listings
```

```python
# price of top 5 with highest npo of reviews
sn.barplot(x = 'neighbourhood', y = 'price', data = sorted_listings)
plt.show()
```
